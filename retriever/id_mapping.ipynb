```python
import json
from pathlib import Path
```

```python
PATH = '../data/qa_clean_data.json'
```

```python
with open(PATH, 'r') as f:
    data = json.load(f)
```

```python
def get_id_mapping(data):
    id_mapping = {}
    for idx, value in enumerate(data):
        id_mapping[str(idx)] = {
                'id': value['id'],
                'question': value['question'],
                'answer': value['answer'],
                'source': value['source'],
                'link': value['link'],
                'tags': value['tags'],
            }
    return id_mapping

```

```python
id_mapping = get_id_mapping(data)
```

```python
id_mapping
```

```python
# save the id_mapping to a file
output_path = Path('../retriever/id_mapping.json')
output_path.parent.mkdir(parents=True, exist_ok=True)
with open(output_path, 'w') as f:
    json.dump(id_mapping, f, indent=4)
```
